```python
# import libraries from standard library and anaconda package
import urllib.request as ur
import re
from bs4 import BeautifulSoup as BS

```

```python
def output(url):
    page = ur.urlopen(url)   
    pagest = str(page)

    soup = BS(page, 'html.parser')
    #extracting content in tags script
    for script in soup(["script"]):      
        script.extract()  
    #extracting content in tags style
    for script in soup(["style"]):      
        script.extract()  
    text = soup.get_text()

    #change into lines and remove leading and trailing space on each
    lines = (line.strip() for line in text.splitlines())
    # break multi-headlines into a line each
    chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
    # drop blank lines
    text = '\n'.join(chunk for chunk in chunks if chunk)
    list=text.split(",")


    k=list[0]#array for list 0

    leng=len(k) # length of file
    date=[] # empty array for dates


    for i in range(leng):
        if k[i]=='R':
            if k[i+1]=='e':
                if k[i+2]=='l':
                    date.append(k[i+15:i+26])
                    print(date)
```
